# Cargar los datos #

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/respuestas_filtered.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
for column in df.columns:
    if column not in ['Sexo', 'Edad']:
        df[column] = df[column].map({1: False, 2: True})
```

```python
X = df.drop('Cancer de pulmon', axis=1)
y = df['Cancer de pulmon']
```

# Importar las librerias correspondientes #

# DecisionTreeClassifier #

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = DecisionTreeClassifier(criterion='entropy', max_depth=30)
```

```python
model1.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
y_pred = model1.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```

# Eliminar valores nulos y atipicos #

```python
test_data.dtypes
```

```python
(test_data.isnull().sum()/(len(test_data)))*100
```

```python
test_data = test_data.dropna()
```

```python
test_data.isna().sum()
```

```python
for i in test_data.columns:
    print(f'{i}: {test_data[i].unique()}')
```

```python
for col in test_data.columns:
  if col == 'Sexo':
    outlier = [v for v in test_data['Sexo'] if not v in ['M', 'F']]
    print(f'{col}: {len(outlier)}')
  elif col not in ['Sexo','Edad', 'Cancer de pulmon']:
    outlier = [v for v in test_data [col] if v not in ['1', '2', 1, 2, 1.0, 2.0]]
    print(f'{col}: {len(outlier)}')
  elif col == 'Edad':
    outlier = [v for v in test_data [col] if v in ['V', '/', '&']]
    print(f'{col}: {len(outlier)}')
```

```python
test_data_filtered = test_data.copy()
for col in test_data_filtered.columns:
  if col not in ['Sexo','Edad']:
    test_data_filtered = test_data_filtered[test_data_filtered[col].isin(['1', '2', 1, 2, 1.0, 2.0])]
```

```python
for i in test_data_filtered.columns:
    print(f'{i}: {test_data_filtered[i].unique()}')
```

```python
test_data_filtered['Sexo'] = test_data_filtered['Sexo'].replace({'F': 2, 'M': 1})
```

```python
test_data_filtered['Sexo'].unique()
```

```python
for col in test_data_filtered.columns:
  test_data_filtered[col] = test_data_filtered[col].astype(int)
```

```python
test_data_filtered.isna().any()
```

```python
test_data_filtered.describe()
```

```python
columns_to_convert = ['Sexo', 'Edad']
for column in test_data_filtered.columns:
    if column not in columns_to_convert:
        test_data_filtered[column] = test_data_filtered[column].map({1: False, 2: True})
```

```python
columnas = test_data_filtered.columns.tolist()
columnas.insert(1, columnas.pop(columnas.index('Edad')))
test_data_filtered = test_data_filtered[columnas]
```
